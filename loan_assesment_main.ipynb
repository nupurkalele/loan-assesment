```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("C:/Users/Nupur/pythonprac/home train data/datasets/loan_data.csv")

#Q1
#information
print(df.info())

#Q2
#compute the mean, max, quantile and min of applicant income and co applicant income seperately using pandas.

new_df=df[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']]
new_df.describe()

#Q3
#plot the histogram of applicant income with 10 bins and co applicant.

df['ApplicantIncome'].shape[0]
plt.hist(df['ApplicantIncome'],bins=50)

#Q4 
#dataframe of males and females whose loan status is yes (approved)
gender_df = df.loc[((df['Gender']=='Male') & (df['Loan_Status']=='Y'))]
gender_df = df.loc[((df['Gender']=='Female') & (df['Loan_Status']=='Y'))]

print(gender_df.reset_index())
print(gender_df.shape)

#Q5 
#df of males whose loan is sactioned
gender_df = df.loc[((df['Gender']=='Male') & (df['Loan_Status']=='Y'))]
print(gender_df.shape)

#Q6
#df of ppl having credit hisrory 1 and loan status as yes
saction_df = df.loc[((df['Credit_History']==1) & (df['Loan_Status']=='Y'))]
print(saction_df.reset_index())
print(saction_df.shape)

#Q7
#pie chart of males and female applicants for which loan is approved.

status_df=df.loc[((df['Loan_Status']=='Y'))]
print(status_df.reset_index())
print(status_df.shape)

status_df["Gender"].value_counts().plot.pie()


#Q7 but using where clause
sorted_df=df.where(df['Loan_Status']=='Y')
print(sorted_df)
sorted_df['Gender'].value_counts().plot.pie()
```
